# **Project Summary -**

This project aimed to explore and understand the evolving structure of Netflixâ€™s global content catalog using unsupervised learning techniques. The dataset, sourced from Flixable (2019), contained over 7,000 titles spanning multiple genres, release years, and countries. Through detailed exploratory analysis and clustering, this study uncovered meaningful patterns in Netflixâ€™s content evolution, genre diversity, and production strategies.

**Understanding the Context**

Over the last decade, Netflix has transitioned from being primarily a movie-streaming platform to a diverse hub for original TV series, documentaries, and international content. This change in strategy is reflected in the dataset, which includes variables such as title, type, country, date_added, release_year, duration, listed_in (genres), and description. By analyzing these factors, the project sought to answer three broad questions:

1. How has Netflixâ€™s content mix evolved over time?

2. What underlying clusters exist among its titles based on textual and numeric features?

3. What business insights can these clusters reveal about Netflixâ€™s global strategy?

**Methodology Overview**

The workflow followed a structured data science pipeline â€” data cleaning, feature engineering, transformation, scaling, and unsupervised modeling.

- Missing values were imputed logically, text columns standardized, and duration fields converted into consistent numeric formats (minutes or seasons).

- New features were engineered such as delay_years (gap between release and addition to Netflix) and num_genres.

- Descriptive text fields like description were transformed using TF-IDF vectorization, converting narrative content into meaningful numerical representations.

- Numeric features were scaled using StandardScaler to bring all values onto a comparable range.

**Modeling and Findings**

Three clustering algorithms were applied: K-Means, Hierarchical Clustering, and DBSCAN, each providing a different perspective on content segmentation.

K-Means Clustering, guided by the elbow method and silhouette analysis, revealed four distinct clusters:

1. Classic Adapted Films (Older, Multi-Genre Dramas)

2. Modern Feature-Length Dramas

3. Short Global Web Shows (Recent, Brief, Modern Themes)

4. Single-Genre Documentaries or Thrillers

These clusters showed strong separation in both numeric (release year, duration, genres) and textual dimensions (key descriptive terms).
Hierarchical Clustering validated these patterns with slightly merged groupings, demonstrating the robustness of the underlying structure.
DBSCAN, a density-based approach, discovered a few compact micro-clusters but classified most data as noise â€” reflecting Netflixâ€™s highly diverse and overlapping catalog rather than clearly isolated content categories.

**Evaluation and Model Selection**

For evaluation, inertia and silhouette scores were used to balance compactness and separability of clusters. Among the models, K-Means achieved the most interpretable and stable clusters, supported by clear feature differentiation and balanced distribution. This made it the final choice for deriving business insights.

**Business Interpretation**

The identified clusters highlight Netflixâ€™s strategic evolution:

Increasing dominance of modern, short-duration series and feature-length dramas post-2015.

A declining proportion of older adapted films and long-format titles.

Rising inclusion of single-genre and documentary-style content, catering to niche audiences.

Such segmentation enables Netflix to personalize recommendations, optimize content acquisition, and strategically plan regional content based on emerging viewer trends. For instance, clusters can guide algorithms to recommend newer, shorter content to younger audiences or promote documentaries in markets showing higher engagement with educational genres.

**Conclusion**

This analysis demonstrates how unsupervised learning can reveal latent structures within large-scale multimedia datasets. Through systematic preprocessing, TF-IDF-based text encoding, and comparative modeling, the study translated unstructured entertainment data into actionable insights. The combination of quantitative features (year, duration, genre count) and qualitative attributes (descriptions, themes) enabled a holistic understanding of Netflixâ€™s catalog evolution.

Overall, the project emphasizes that clustering is not only a technical exercise but a tool for strategic decision-making â€” helping streaming platforms like Netflix refine personalization, balance genre diversity, and anticipate future content trends. This end-to-end approach, from data engineering to model explainability, provides a scalable blueprint for other OTT platforms seeking to derive business intelligence from their content ecosystems.

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Create a working copy
df = df_raw.copy()

# -------------------------------
# 1ï¸ Handle Missing & Placeholder Values
# -------------------------------
placeholders = ['no data', 'n/a', 'na', 'unknown', 'none', '-', 'not specified']
df = df.replace(placeholders, pd.NA)

text_cols = ['title', 'director', 'cast', 'listed_in', 'country', 'rating']
for col in text_cols:
    df[col] = df[col].astype(str).str.strip().str.lower()

# -------------------------------
# 2ï¸ Convert Date Column to Datetime
# -------------------------------
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# -------------------------------
# 3ï¸ Handle Remaining Missing Values
# -------------------------------
# Drop rows with missing date_added or rating
df = df.dropna(subset=['date_added', 'rating'])

# Fill missing values in text-based columns
df['director'] = df['director'].fillna('unknown')
df['cast'] = df['cast'].fillna('unknown')
df['country'] = df['country'].fillna('unknown')

# Reset index
df = df.reset_index(drop=True)

# ------------------------------
# 4ï¸ Process Duration Column
# -------------------------------
df['duration'] = df['duration'].fillna('')

# Extract numeric part
df['duration_int'] = df['duration'].str.extract(r'(\d+)')
df['duration_int'] = pd.to_numeric(df['duration_int'], errors='coerce')

# Extract and normalize duration type
df['duration_type'] = df['duration'].str.extract(r'([a-zA-Z]+)')
df['duration_type'] = df['duration_type'].fillna('').str.lower()
df['duration_type'] = df['duration_type'].replace({'seasons': 'season'})

# -------------------------------
# 5 Final Cleaning Check
# -------------------------------
print(" Missing Values after cleaning:\n", df.isnull().sum())
print("\n Shape after cleaning:", df.shape)

# # Save cleaned data for EDA
# df.to_csv("/content/drive/MyDrive/netflix_cleaned.csv", index=False)
# print("\n Cleaned file saved as 'netflix_cleaned.csv' in your Drive.")
````

### What all manipulations have you done and insights you found?

**Data wrangling summary**

- Replaced common placeholders (`'na','no data','-'`) with NaN to standardize missing values.
- Converted `date_added` to datetime for time-based analysis.
- Filled sparse metadata: `director`, `cast`, `country` with 'Unknown' to retain records.
- Parsed `duration` into `duration_int` and `duration_type` (minutes vs season).
- Created numeric features: `cast_count`, `genre_count`, `content_age`.
- These features enable numeric analysis, clustering validation, and allow combining textual TF-IDF features with metadata.

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Null Hypothesis (Hâ‚€):
There is no significant difference in the average duration between Movies and TV Shows on Netflix.

Alternative Hypothesis (Hâ‚):
There is a significant difference in the average duration between Movies and TV Shows.

#### 2. Perform an appropriate statistical test.

````python
# Ensure correct mapping for type
df['type'] = df['type'].replace({'Movie': 0, 'TV Show': 1})

# Filter valid numeric durations
df = df[df['duration_int'].notnull() & (df['duration_int'] > 0)]

# Separate groups again
movies = df[df['type'] == 0]['duration_int']
shows = df[df['type'] == 1]['duration_int']

# Check descriptive stats before test
print("Movies mean duration:", movies.mean())
print("TV Shows mean duration:", shows.mean())
print("Movies count:", len(movies), " | TV Shows count:", len(shows))

# Run independent t-test

t_stat, p_val = ttest_ind(movies, shows, equal_var=False)

print("\nT-statistic:", t_stat)
print("P-value:", p_val)

if p_val < 0.05:
    print("\n Reject Hâ‚€: There IS a significant difference in duration between Movies and TV Shows.")
else:
    print("\n Fail to reject Hâ‚€: No significant difference in duration between Movies and TV Shows.")
````

Independent Samples t-test.

The t-test is the most suitable test when:

The dependent variable (duration) is numeric and approximately normal,

The independent variable (content type) has two categories, and

The samples are independent of each other.

### 2. Handling Outliers

````python
# 1ï¸ Select numerical columns for outlier detection
numeric_cols = ['release_year', 'duration_int', 'delay_years']

# 2ï¸ Define function to detect outliers using IQR
def detect_outliers_iqr(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    return {
        "col": col,
        "lower_bound": round(lower_bound, 2),
        "upper_bound": round(upper_bound, 2),
        "below_lower": len(df[df[col] < lower_bound]),
        "above_upper": len(df[df[col] > upper_bound]),
        "total_outliers": len(outliers)
    }

# 3ï¸ Detect outliers
outlier_summary = [detect_outliers_iqr(df, col) for col in numeric_cols]

# 4ï¸ Display summary

outlier_summary_df = pd.DataFrame(outlier_summary)
print("=====  Outlier Summary (Before Treatment) =====")
display(outlier_summary_df)

# 5ï¸ Handle outliers â€” capping (Winsorization)
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)

print("\n Outlier treatment completed successfully!")
print(df[numeric_cols].describe())

plt.figure(figsize=(12, 6))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(1, 3, i)
    sns.boxplot(x=df[col], color='teal')
    plt.title(f"Boxplot After Outlier Treatment\n({col})", fontsize=11)
plt.tight_layout()
plt.show()
````

##### What all outlier treatment techniques have you used and why did you use those techniques?

The Interquartile Range (IQR) Method was used to detect outliers in numeric columns â€” release_year, duration_int, and delay_years.

Outliers were defined as points outside Q1 âˆ’ 1.5Ã—IQR and Q3 + 1.5Ã—IQR.

Instead of dropping records (which risks losing important data), Winsorization was applied â€” capping outlier values at the lower and upper bounds.

This ensures that data distributions remain consistent and robust for further statistical analysis.

Reason for using IQR method:
Itâ€™s robust to skewed data and unaffected by extreme values, making it suitable for real-world datasets like Netflix titles that span many years and durations.

For text vectorization, the TF-IDF (Term Frequencyâ€“Inverse Document Frequency) technique was used.

It was chosen because:

It reduces the influence of common words (like â€œtheâ€, â€œandâ€) that appear frequently but carry little meaning.

It emphasizes rare yet informative terms, improving the modelâ€™s ability to distinguish between texts.

Itâ€™s more effective than simple Count Vectorization, which treats all words equally.

It provides a sparse, interpretable representation suitable for clustering, classification, or similarity analysis tasks.

Hence, TF-IDF is ideal for this projectâ€™s textual fields (like description) to numerically represent content importance while preserving contextual distinctiveness.

### ML Model - 1:  K-Means Clustering Implementation

**1. Objective:**

To group similar Netflix titles (movies and shows) based on both textual (TF-IDF/PCA) and numerical (scaled) features.

````python

# -----------------------------
# Step 1: Prepare data for clustering
# -----------------------------
# Use the combined features from PCA + scaled numeric data
X = combined_features

# -----------------------------
# Step 2: Elbow & Silhouette Method
# -----------------------------
K = range(2, 11)   # Try between 2 to 10 clusters
inertia = []
silhouette = []

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)
    silhouette.append(silhouette_score(X, kmeans.labels_))

# -----------------------------
# Step 3: Print results
# -----------------------------
print("K | Inertia | Silhouette Score")
print("-" * 40)
for k, i, s in zip(K, inertia, silhouette):
    print(f"{k:<2} | {i:>10.2f} | {s:>8.4f}")

# -----------------------------
# Step 4: Plot Elbow Method
# -----------------------------
plt.figure(figsize=(8,4))
plt.plot(K, inertia, 'bo-', linewidth=2)
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()

# -----------------------------
# Step 5: Plot Silhouette Scores
# -----------------------------
plt.figure(figsize=(8,4))
plt.plot(K, silhouette, 'ro-', linewidth=2)
plt.title('Silhouette Scores for Different k')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Silhouette Score')
plt.grid(True)
plt.show()
````

### ML Model - 3 DBSCAN CLUSTERING

**Model Overview**

**Purpose:**
To identify clusters of arbitrary shape and size and handle outliers (noise) that K-Means or Hierarchical methods may force into clusters.

**Algorithm Type:** Density-Based Unsupervised Clustering
Core Parameters:

 - eps â€“ maximum distance between two points to be considered neighbors.

 - min_samples â€“ minimum number of points to form a dense region.

````python
# --- MODEL 3: DBSCAN CLUSTERING ---

# from sklearn.cluster import DBSCAN
# from sklearn.preprocessing import StandardScaler
# from sklearn.metrics import silhouette_score
# import numpy as np

# --- Combine numeric + text features as before ---
# (Assuming you already have: combined_features, df, tfidf_matrix, vectorizer, etc.)
# If not, rebuild combined_features same as in Model 1.

# --- Step 1: Standardize combined features ---
scaler = StandardScaler()
scaled_features = scaler.fit_transform(combined_features)

# --- Step 2: Apply DBSCAN ---
dbscan = DBSCAN(eps=1.8, min_samples=5, metric='euclidean')  # tune eps if needed
df['db_cluster'] = dbscan.fit_predict(scaled_features)

# --- Step 3: Evaluate clustering performance ---
unique_clusters = np.unique(df['db_cluster'])
n_clusters = len(unique_clusters[unique_clusters != -1])  # exclude noise (-1)
noise_points = np.sum(df['db_cluster'] == -1)

print(f"Number of clusters (excluding noise): {n_clusters}")
print(f"Number of noise points: {noise_points}")

# Compute silhouette only if thereâ€™s more than 1 cluster
if n_clusters > 1:
    sil_score = silhouette_score(scaled_features[df['db_cluster'] != -1],
                                 df['db_cluster'][df['db_cluster'] != -1])
    print(f"Silhouette Score: {sil_score:.3f}")
else:
    print("Silhouette score not applicable (only one cluster found).")

# --- Step 4: Cluster Distribution ---
print("\nCluster Distribution:")
print(df['db_cluster'].value_counts())

# --- Step 5: Numeric feature summary per cluster ---
numeric_summary = df.groupby('db_cluster')[['release_year', 'duration_int', 'num_genres', 'delay_years']].mean()
print("\nNumeric Feature Summary by DBSCAN Cluster:")
print(numeric_summary)

# --- Step 6: Top Terms per DBSCAN Cluster (Fixed Version) ---
terms = vectorizer.get_feature_names_out()

for i in range(n_clusters):
    cluster_indices = np.where(df['db_cluster'].values == i)[0]  # indices for this cluster

    if len(cluster_indices) == 0:
        print(f"\nCluster {i}: No samples found.")
        continue

    cluster_mean = tfidf_matrix[cluster_indices].mean(axis=0)
    top_indices = np.array(cluster_mean).ravel().argsort()[-10:][::-1]
    top_terms = [terms[j] for j in top_indices]

    print(f"\nCluster {i} Top Terms: {', '.join(top_terms)}")

# --- Step 7: Example titles per cluster ---
print("\nExample Titles with DBSCAN Cluster Labels:")
print(df[['title', 'db_cluster']].head(10))
````
